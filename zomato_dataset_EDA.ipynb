```python
df_country = pd.read_excel('/content/Country-Code.xlsx')
df_country.head()
```

```python
# merging of two data frame
final_df = pd.merge(df,df_country,on='Country Code', how='left')
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df.columns
```

```python
country_name = final_df.Country.value_counts().index
country_name
```

```python
country_values = final_df.Country.value_counts().values
```

#**Data Visualisation**

#**PIE chart of Top 3 countries that are using zomato**

```python
plt.pie(country_values[:3], labels = country_name[:3], autopct='%1.3f%%')
plt.show()
```

```python
final_df.head()
```

#**Making new data frame using grouping of [ Aggregate rating, Rating color, Rating text ]**

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head()
```

#**OBSERVATIONS**
*   when ratings are betwn 4.5 - 4.9 --> Excellent
*   when ratings are betwn 4.1 - 4.4 --> Very good
*   when ratings are betwn 3.5 - 3.9 --> Good
*   when ratings are betwn 2.5 - 3.4 --> Average
*   when ratings are betwn 1.8 - 2.4 --> Poor
*   when ratings are betwn 0.0 --> Not RAted





```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(data = ratings , x='Aggregate rating', y='Rating Count', palette='Spectral')
plt.show()
```

#**OBSERVATIONS**
*   Not rated count is very high
*   Maximum ratings are between 2.5 to 3.9
*   This concludes that if ratings are missing than we would consider it as in average rating since maximum ratings lies in average






```python
sns.barplot(data =ratings, x='Aggregate rating', y='Rating Count', hue = 'Rating color', palette=['black','red','orange','yellow','green','brown'])
plt.show()
```

```python
sns.countplot(data = ratings, x='Rating color', palette=['black','red','orange','yellow','green','brown'])
plt.show()
```
